# DataFrame 'GroupBy' Method

Using "group by" method we can:
• Split the data into groups based on some criteria
• Calculate statistics (or apply a function) to each group
• Similar to dplyr() function in R

```python
##Group data using rank
import pandas as pd 
df = pd.read_csv("http://rcs.bu.edu/examples/python/data_analysis/Salaries.csv")

df_rank = df.groupby (['rank'])

#Calculate mean value for each numeric column per each group

df_rank.mean()
```

```python
#Calculate mean salary for each professor rank:

df.groupby('rank')[['salary']].mean()

#Note: If single brackets are used to specify the column (e.g. salary), then the output is Pandas Series object.
#When double brackets are used the output is a Data Frame
```

```python
#by default the group keys are sorted during the groupby operation. You may
#want to pass sort=False for potential speedup
#Calculate mean salary for each professor rank:
df.groupby(['rank'], sort = False )[['salary']].mean()
```

# DataFrame Slicing

There are a number of ways to subset the Data Frame:
• one or more columns
• one or more rows
• a subset of rows and columns
Rows and columns can be selected by their position or label 

```python
#When selecting one column, it is possible to use single set of brackets, but the
#resulting object will be a Series (not a DataFrame):

#Select column salary:

df['salary']

#When we need to select more than one column and/or make the output to be a
#DataFrame, we should use double brackets:

df[['rank' , 'salary']]
```

```python
#If we need to select a range of rows, we can specify the range using ":" 

#Select rows by their position:

df[0:10]

#Notice that the first row has a position 0, and the last value in the range is omitted:
#So for 0:10 range the first 10 rows are returned with the positions starting with 0
#and ending with 9
```

# Data Frames: method loc

If we need to select a range of rows, using their labels we can use method loc:


```python
#select rows by there labels
import pandas as pd 
df = pd.read_csv("http://rcs.bu.edu/examples/python/data_analysis/Salaries.csv")
df.loc[0:10, ['rank','sex','salary']]
```

### .loc method 
--Only uses labels
--Can select rows and columns simultaneously
--Selection can be a single label, a list of labels or a slice of labels
--Put a comma between row and column selections
--is inclusive of last integer when selecting a row 

# Dataframe :Method iloc

The .iloc indexer is very similar to .loc but only uses integer locations to make its selections. The word .iloc itself stands for integer location so that should help with remember what it does.

```python
#Select row using there postions or indexes
#Selecting a single row with .iloc
df.iloc[3]
```

```python
#Selecting a multiple  row with .iloc

df.iloc[[2,3,4]] # remember not to do as df.iloc[2,3,4]
```

```python
#Use slice notation to select a range of rows with iloc

df.iloc[3:5]
```

```python
#Select 3rd position until end

df.iloc[3:]
```

```python
#select 3rd positon to end by 2

df.iloc[3::2]
```

### Selecting rows and columns simultaneously by iloc method

Just like with .iloc any combination of a single integer, lists of integers or slices can be used to select rows and columns simultaneously. Just remember to separate the selections with a "comma".

```python
#Select two rows and two columns:

df.iloc[[2,3] , [0,3]]
```

```python
#Select a slice of the rows and two columns:

df.iloc[0:3,[0,3]]
```

```python
#select slices for both rows and columns

df.iloc[0:3 , 0:3]
```

```python
#Select a single row and column

df.iloc[0,2]
```

```python
#Select all the rows and a single column

df.iloc[:,1]
```

```python
#select last row 

df.iloc[-1]
```

# Dataframe : Sorting 

We can sort the data by a value in the column. By default the sorting will occur in
ascending order and a new data frame is return. 

```python
# Create a new data frame from the original sorted by the column Salary

df_sorted = df.sort_values(by = 'service')
df_sorted.head()
```

```python
#We can sort the data using 2 or more columns

df_sorted = df.sort_values(by = ['service','salary'],ascending = [True , False])
df_sorted.head(10)
```
